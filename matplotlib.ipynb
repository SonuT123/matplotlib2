```python
"""Matplotlib is a widely-used 2D plotting library for Python.
     It provides a convenient way to create a variety of static, animated,
     and interactive visualizations in Python.
     Matplotlib allows you to create various types of plots, charts, and graphs to visualize data and patterns.
     
   Matplotlib is used for:-
   
   >Creating visualizations of data in the form of plots, charts, and graphs.
   >Exploring and analyzing data trends, relationships, and patterns.
   >Communicating complex data insights in a more understandable and visual manner.
   >Generating publication-quality figures for research papers, presentations, and reports.
   
   
   
   Five plots that can be plotted using the Pyplot module of Matplotlib:-
   
   1.Line Plot: Used to represent data points with connected lines, often showing the trend or 
     relationship between data points.
     
   2.Bar Plot: Displays rectangular bars to represent data, often used to compare different categories or groups.
   
   3.Histogram: Represents the distribution of data values in a continuous range by dividing data into bins and 
     showing the frequency of values within each bin.
     
   4.Scatter Plot: Represents individual data points as dots on a 2D plane, used to display the relationship
     between two variables.
     
   5.Pie Chart: Represents data as a circular chart divided into sectors, where each sector represents 
     a proportion of the whole data.   """
     
```

```python
"""A scatter plot is a type of data visualization that displays individual data points as dots on a 2D plane.
    It is particularly useful for showing the relationship or correlation between two variables.
    Each dot represents a single data point, where the x-coordinate represents one variable
    and the y-coordinate represents another variable."""
    

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))


plt.scatter(x, y, color='blue', label='Data Points')


plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.legend()


plt.show()
```

```python
"""The subplot() function in Matplotlib is used to create multiple plots within a single figure.
   It allows you to create a grid of subplots, where you can place different plots in different cells of the grid.
   This is helpful when you want to visualize multiple plots side by side or in a grid layout."""


import numpy as np
import matplotlib.pyplot as plt


x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')

plt.legend()


plt.tight_layout()

plt.show()
```

```python
"""A bar plot is a type of data visualization that uses rectangular bars to represent data values.
     It is particularly useful for comparing and displaying the data values of different categories or groups.
     The length of each bar is proportional to the value it represents.
     
    A horizontal bar plot is similar to a bar plot, but the bars are oriented horizontally.  """


import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='blue')
plt.title('Bar Plot')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='green')
plt.title('Horizontal Bar Plot')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.show()    
```

```python
""" A box plot, also known as a box-and-whisker plot, is a data visualization technique used to display 
    the distribution of a dataset. It shows the median, quartiles, and potential outliers of the data.
    The box in the plot represents the interquartile range (IQR), which contains the middle 50% of the data.


    The "whiskers" extend to show the data within a certain range, while points outside the whiskers are considered 
   potential outliers.
   Box plots are particularly useful for comparing the distribution of data across different groups or
   categories and identifying skewness, central tendency, spread, and potential outliers in the data."""
   
   
   
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()    
```
